```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pickle
import seaborn as sns

from sklearn.utils.class_weight import compute_class_weight
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import feature_extraction, linear_model, model_selection, preprocessing
from sklearn.naive_bayes import MultinomialNB

from sklearn import preprocessing

from sklearn.cluster import KMeans

%matplotlib inline
```

```python
df = pd.read_csv('/home/arch/.datasets/calhouse/housing.csv')
```

```python
dff = df[['longitude', 'latitude', 'median_house_value']]
```

```python
sns.scatterplot(data = dff, x = 'longitude', y = 'latitude', hue = 'median_house_value')
```

```python
kmeans = KMeans(n_clusters = 3, random_state = 0, n_init='auto')
kmeans.fit(dff)
```

```python
sns.scatterplot(data = dff, x = 'longitude', y = 'latitude', hue=kmeans.predict(dff))
```
